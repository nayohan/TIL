##### np.linspace()

```python
arr7 = np.logspace(20, 100, 11)
print(arr7)
```

##### np.random.normal(mean, std, size=(m,n))
* 정규분포확률밀도함수에서 표본 추출
* 10000개의 표본을 평균 0, 표준편차1에 해당하는 히스토그램(100개구간)으로 표현
* 평균과 표준편차를 변경하고 싶다면 사용

```python
arr8 = np.random.normal(0, 1, 10000)
print(arr8.shape)
plt.hist(arr8, bins=100)
plt.show()
```

##### np.random.rand(d0, d1, ..., dn)
* 0에서 1사이의 숫자중 균등분포 형상으로 표본 추출
* 고르게 분포하는 난수를 생성할때 사용

```python
arr8 = np.random.rand(10000, 1)
print(arr8.shape)
plt.hist(arr8, bins=100)
plt.show()
```

##### np.random.randn(d0, d1, ..., dn)
* 표준정규분포에서 표본 추출
* 표준정규분포 : 평균0, 표준편차1인 정규분포

```python
arr8 = np.random.randn(10000,1)
plt.hist(arr8, bins=100)
plt.show()
```

##### np.random.randint(low, high, size, dtype)
* 지정된 범위(low->high)까지에서 size만큼 정수표본 추출
* 10에서 100까지 1000개의 표본을 균등하게 추출

```python
arr8 = np.random.randint(10,100, 10000, 'int32')
plt.hist(arr8, bins=90)
plt.show()
```
